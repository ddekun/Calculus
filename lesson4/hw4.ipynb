### 2) Построить график функции:
$a) y = x^2+4x+3$

```python
x = np.linspace(-40, 40, 10000)
y = x ** 2 + 4 * x + 3
plt.figure(1)
plt.title('График функции y = x^2 + 4x + 3')
plt.ylabel('Ось y')
plt.xlabel('Ось x')
plt.grid()
plt.axis([-10, 5, -2, 10]) 
plt.plot(x, y)
plt.show()
```
